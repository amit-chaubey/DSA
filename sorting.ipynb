```python
def bubbleSort(a):
  for i in range(len(a)):
    for j in range(len(a)-i-1):
      if a[j] > a[j+1]:
        a[j],a[j+1] = a[j+1], a[j]
  return a


l = [3,2,1,8,4]
print(bubbleSort(l))


```

```python
def selectionSort(s):
  for i in range(len(s)):
    curr_idx = i
    for j in range(i, len(s)):
      if s[curr_idx] > s[j]:
        curr_idx = j
    s[i], s[curr_idx] = s[curr_idx], s[i]
  return s


s = [3,2,1,5,4]
print(selectionSort(s))
```
